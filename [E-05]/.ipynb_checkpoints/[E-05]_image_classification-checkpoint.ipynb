```python
import tensorflow as tf
from tensorflow import keras

import matplotlib.pyplot as plt
import numpy as np

from PIL import Image
import glob
import os

print(tf.__version__)
print(np.__version__)
```

### 이미지 데이터 로드

- 해당 이미지 데이터의 사이즈는 300X300으로 MNIST 데이터셋에 비해 높은 해상도를 가져 모델의 성능이 높게 도출됨
- 이번 노드에서는 데이터 전처리 및 하이퍼파라미터 조정에 따른 모델의 성능을 비교하기 위해 해상도를 128X128로 resize 한 후 분석 진행

```python
# 이미지 사이즈 300x300

image_dir_path = "./Rock Paper Scissors.v1-raw-300x300.folder (2)/train"
val_image_dir_path = "./Rock Paper Scissors.v1-raw-300x300.folder (2)/valid"
test_image_dir_path = "./Rock Paper Scissors.v1-raw-300x300.folder (2)/test"
```

### 데이터 resize 메서드

```python
def resize_data(img_path):  # 가위바위보 이미지 개수 총합에 주의하세요.
    
    idx=0
    # 파일마다 모두 128x128 사이즈로 바꾸어 저장합니다.
    target_size=(128,128)
    for file in glob.iglob(img_path+'/scissor/*.jpg'):
        old_img=Image.open(file)
        new_img=old_img.resize(target_size,Image.ANTIALIAS)
        new_img.save(file, "JPEG")

    for file in glob.iglob(img_path+'/rock/*.jpg'):
        old_img=Image.open(file)
        new_img=old_img.resize(target_size,Image.ANTIALIAS)
        new_img.save(file, "JPEG")
    
    for file in glob.iglob(img_path+'/paper/*.jpg'):
        old_img=Image.open(file)
        new_img=old_img.resize(target_size,Image.ANTIALIAS)
        new_img.save(file, "JPEG")
```

### 데이터 로드 메서드

```python
def load_data(img_path, number_of_data=2520):  # 가위바위보 이미지 개수 총합에 주의하세요.
    # 가위 : 0, 바위 : 1, 보 : 2
    img_size=128
    color=3
    #이미지 데이터와 라벨(가위 : 0, 바위 : 1, 보 : 2) 데이터를 담을 행렬(matrix) 영역을 생성합니다.
    imgs=np.zeros(number_of_data*img_size*img_size*color,dtype=np.int32).reshape(number_of_data,img_size,img_size,color)
    labels=np.zeros(number_of_data,dtype=np.int32)

    idx=0
    for file in glob.iglob(img_path+'/scissor/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=0   # 가위 : 0
        idx=idx+1

    for file in glob.iglob(img_path+'/rock/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=1   # 바위 : 1
        idx=idx+1  
    
    for file in glob.iglob(img_path+'/paper/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=2   # 보 : 2
        idx=idx+1
        
    print("이미지 개수는", idx,"입니다.")
    return imgs, labels
```

### 데이터 전처리

참고: https://keraskorea.github.io/posts/2018-10-24-little_data_powerful_model/

#### 1) 정규화

```python
resize_data(image_dir_path)
(x_train, y_train)=load_data(image_dir_path)
x_train_norm = x_train/255.0   # 입력은 0~1 사이의 값으로 정규화

resize_data(val_image_dir_path)
(x_val, y_val)= load_data(val_image_dir_path, number_of_data=372)
x_val_norm = x_val/255.0

resize_data(test_image_dir_path)
(x_test, y_test)= load_data(test_image_dir_path, number_of_data=33)
x_test_norm = x_test/255.0

print("x_train shape: {}".format(x_train.shape))
print("y_train shape: {}".format(y_train.shape))

print("x_valid shape: {}".format(x_val.shape))
print("y_valid shape: {}".format(y_val.shape))

print("x_test shape: {}".format(x_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

- 기존에 받아온 데이터가 컬러 데이터이기 때문에 reshape를 통한 채널 추가를 해줄 필요가 없지만 일단 진행합니다.

```python
print(f"Before Reshape - x_train_norm shape: {x_train_norm.shape}")
print(f"Before Reshape - x_test_norm shape: {x_test_norm.shape}")

x_train_reshaped=x_train_norm.reshape(-1, 128, 128, 3)  # 데이터갯수에 -1을 쓰면 reshape시 자동계산됩니다.
x_val_reshaped=x_val_norm.reshape(-1, 128, 128, 3) 
x_test_reshaped=x_test_norm.reshape(-1, 128, 128, 3)

print(f"After Reshape - x_train_reshaped shape: {x_train_reshaped.shape}")
print(f"After Reshape - x_valid_reshaped shape: {x_val_reshaped.shape}")
print(f"After Reshape - x_test_reshaped shape: {x_test_reshaped.shape}")
```

### 모델링

#### 1번 모델
1) 하이퍼파라미터 설정 및 모델 생성, 훈련, 테스트 결과

```python
# 하이퍼파라미터 설정
n_channel_1=16
n_channel_2=32
n_dense=32
n_train_epoch=20

model=keras.models.Sequential()
model.add(keras.layers.Conv2D(n_channel_1, (3,3), activation='relu', input_shape=(128,128,3)))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Conv2D(n_channel_2, (3,3), activation='relu'))
model.add(keras.layers.MaxPooling2D((2,2)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(n_dense, activation='relu'))
model.add(keras.layers.Dense(3, activation='softmax'))

model.summary()
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

# 모델 훈련
hist = model.fit(x_train_reshaped, y_train, epochs=n_train_epoch, validation_data=(x_val_reshaped, y_val))

# 모델 시험
test_loss, test_accuracy = model.evaluate(x_test_reshaped, y_test, verbose=2)
print(f"test_loss: {test_loss} ")
print(f"test_accuracy: {test_accuracy}")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')

acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

2) 추론 결과 시각화

```python
predicted_result = model.predict(x_test_reshaped)  # model이 추론한 확률값. 
predicted_labels = np.argmax(predicted_result, axis=1)

idx=0  #1번째 x_test를 살펴보자. 
print('model.predict() 결과 : ', predicted_result[idx])
print('model이 추론한 가장 가능성이 높은 결과 : ', predicted_labels[idx])
print('실제 데이터의 라벨 : ', y_test[idx])
```

```python
import random
wrong_predict_list=[]
for i, _ in enumerate(predicted_labels):
    # i번째 test_labels과 y_test이 다른 경우만 모아 봅시다. 
    if predicted_labels[i] != y_test[i]:
        wrong_predict_list.append(i)

# wrong_predict_list 에서 랜덤하게 5개만 뽑아봅시다.
samples = random.choices(population=wrong_predict_list, k=5)

for n in samples:
    print("예측확률분포: " + str(predicted_result[n]))
    print("라벨: " + str(y_test[n]) + ", 예측결과: " + str(predicted_labels[n]))
    plt.imshow(x_test[n], cmap=plt.cm.binary)
    plt.show()
```

#### 2번 모델

하이퍼 파라미터 변경

- n_dense = 100
- n_channel_1 = 32
- n_channel_2 = 64

```python
n_channel_1=32
n_channel_2=64
n_dense= 100
n_train_epoch=20

model=keras.models.Sequential()
model.add(keras.layers.Conv2D(n_channel_1, (3,3), activation='relu', input_shape=(128,128,3)))
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Conv2D(n_channel_2, (3,3), activation='relu'))
model.add(keras.layers.MaxPooling2D((2,2)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(n_dense, activation='relu'))
model.add(keras.layers.Dense(3, activation='softmax'))

model.summary()
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

# 모델 훈련
hist = model.fit(x_train_reshaped, y_train, epochs=n_train_epoch, validation_data=(x_val_reshaped, y_val))

# 모델 시험
test_loss, test_accuracy = model.evaluate(x_test_reshaped, y_test, verbose=2)
print(f"test_loss: {test_loss} ")
print(f"test_accuracy: {test_accuracy}")
```

```python
fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')

acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
```

2) 추론 결과 시각화

```python
predicted_result = model.predict(x_test_reshaped)  # model이 추론한 확률값. 
predicted_labels = np.argmax(predicted_result, axis=1)

idx=0  #1번째 x_test를 살펴보자. 
print('model.predict() 결과 : ', predicted_result[idx])
print('model이 추론한 가장 가능성이 높은 결과 : ', predicted_labels[idx])
print('실제 데이터의 라벨 : ', y_test[idx])
```

```python
import random
wrong_predict_list=[]
for i, _ in enumerate(predicted_labels):
    # i번째 test_labels과 y_test이 다른 경우만 모아 봅시다. 
    if predicted_labels[i] != y_test[i]:
        wrong_predict_list.append(i)

# wrong_predict_list 에서 랜덤하게 5개만 뽑아봅시다.
samples = random.choices(population=wrong_predict_list, k=5)

for n in samples:
    print("예측확률분포: " + str(predicted_result[n]))
    print("라벨: " + str(y_test[n]) + ", 예측결과: " + str(predicted_labels[n]))
    plt.imshow(x_test[n], cmap=plt.cm.binary)
    plt.show()
```

검지와 중지가 길어서 혼선을 빚는 분들이 있는 것 같다. 

### 회고

- 이미지 데이터셋의 구조를 알고, 변환해서 적절하게 모델링을 하는 것이 중요한 것 같다.

- 사용한 데이터셋의 퀄리티가 좋아서 모델 예측이 수월하게 진행된 것 같다.

- train 데이터에 대한 overfitting이 있었고, test 데이터셋의 양을 좀 더 늘려서 진행하는 것이 보다 확실한 모델의 성능 예측에 도움이 될 것 같다.

- ImageDataGenerator를 사용해서 이미지 변형 후 train data에 대한 overfitting을 줄이는 것도 좋을 것 같다.
